```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("test_output/root.tsv", sep='\t', comment='#')
# Plot a histogram (bar chart with bins)
plt.figure(figsize=(10, 5))
plt.hist(data['ttime'], bins=20, color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel("Response Time (ms)")
plt.ylabel("Number of Requests")
plt.title("Distribution of Server Response Times")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the benchmark results for each endpoint
root_data = pd.read_csv("test_output/root.tsv", sep='\t', comment='#')
update_data = pd.read_csv("test_output/update.tsv", sep='\t', comment='#')
get_list_data = pd.read_csv("test_output/get_list.tsv", sep='\t', comment='#')

import seaborn as sns

df = pd.DataFrame({
    "Endpoint": ["/"] * len(root_data)  + ["/get_list"] * len(get_list_data),
    "Response Time (ms)": list(root_data['ttime']) +  list(get_list_data['ttime'])
})

# Create a box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x="Endpoint", y="Response Time (ms)", data=df, palette=["blue", "green", "red"], showfliers=False)

# Labels and title
plt.xlabel("API Endpoints")
plt.ylabel("Response Time (ms)")
plt.title("Response Time Distribution Across Endpoints")

plt.show()
```
